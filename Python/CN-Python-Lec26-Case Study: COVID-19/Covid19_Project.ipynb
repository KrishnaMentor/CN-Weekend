```python
# Melt -> Unpivot to the DataFrame
# pandas.melt
# pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value')
all_cols = confirmed_cases.columns
all_cols
```

```python
id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']
date_cols = [c for c in all_cols if c not in id_vars] # List Comprehension
date_cols
```

```python
date_cols = []
for c in all_cols:
    if c not in id_vars: 
        date_cols.append(c)
date_cols
```

```python
id_vars = confirmed_cases.iloc[:,:4].columns
id_vars
```

```python
date_cols = confirmed_cases.iloc[:,4:].columns
date_cols
```

```python
confirmed_long = confirmed_cases.melt(
    id_vars = id_vars, 
    value_vars = date_cols, 
    var_name = "Date", 
    value_name='Confirmed'
)
confirmed_long
```

```python
deaths_long = deaths.melt(
    id_vars = id_vars, 
    value_vars = date_cols, 
    var_name = "Date", 
    value_name='Deaths'
)
deaths_long
```

```python
recovered_long = recovered_cases.melt(
    id_vars = id_vars, 
    value_vars = date_cols, 
    var_name = "Date", 
    value_name='Recovered'
)
recovered_long
```

```python
# Identifying the missing values
confirmed_long['Province/State'].isna().sum()
```

```python
confirmed_long['Province/State'].value_counts(dropna = False)
```

```python
# Confirmed_long -> [Data Cleaning and Handling Missing Values]
confirmed_long['Province/State'] = confirmed_long['Province/State'].fillna('All Provinces')
confirmed_long['Lat'] = pd.to_numeric(confirmed_long['Lat'].fillna(0))
confirmed_long['Long'] = pd.to_numeric(confirmed_long['Long'].fillna(0))
confirmed_long['Date'] = pd.to_datetime(confirmed_long['Date'] , format = "%m/%d/%y" , errors = 'coerce')
confirmed_long.info()
```

```python
confirmed_long.head()
```

```python
deaths_long.info()
```

```python
# Deaths_long-> [Data Cleaning and Handling Missing Values]
deaths_long['Province/State'] = deaths_long['Province/State'].fillna('All Provinces')
deaths_long['Lat'] = pd.to_numeric(deaths_long['Lat'].fillna(0))
deaths_long['Long'] = pd.to_numeric(deaths_long['Long'].fillna(0))
deaths_long['Date'] = pd.to_datetime(deaths_long['Date'] , format = "%m/%d/%y" , errors = 'coerce')
deaths_long['Deaths'] = pd.to_numeric(deaths_long['Deaths'].fillna(0).astype('int64'))
deaths_long.info()
```

```python
deaths_long.head()
```

```python
recovered_long.info()
```

```python
# Recovered_Long -> [Data Cleaning and Handling Missing Values]
recovered_long['Province/State'] = recovered_long['Province/State'].fillna('All Provinces')
recovered_long['Lat'] = pd.to_numeric(recovered_long['Lat'].fillna(0))
recovered_long['Long'] = pd.to_numeric(recovered_long['Long'].fillna(0))
recovered_long['Date'] = pd.to_datetime(recovered_long['Date'] , format = "%m/%d/%y" , errors = 'coerce')
recovered_long['Recovered'] = pd.to_numeric(recovered_long['Recovered'].fillna(0).astype('int64'))
recovered_long.info()
```

```python
# GroupBY + Aggregations [Feature Selection]
# cd stands for [Country + Date]
confirmed_cd = (
    confirmed_long.groupby(['Country/Region' , 'Date'])['Confirmed'].sum().reset_index()
)
confirmed_cd
```

```python
confirmed_cd[confirmed_cd['Country/Region'] == 'India'].tail(50)
```

```python
# GroupBY + Aggregations [Feature Selection]
# cd stands for [Country + Date]
deaths_cd = (
    deaths_long.groupby(['Country/Region' , 'Date'])['Deaths'].sum().reset_index()
)
deaths_cd
```

```python
# GroupBY + Aggregations [Feature Selection]
# cd stands for [Country + Date]
recovered_cd = (
    recovered_long.groupby(['Country/Region' , 'Date'])['Recovered'].sum().reset_index()
)
recovered_cd
```

```python
# pandas.merge
# pandas.merge(left, right, how='inner', on=None)
merged = (confirmed_cd
          .merge(deaths_cd , on = ['Country/Region','Date'], how = 'inner')
          .merge(recovered_cd , on = ['Country/Region','Date'], how = 'inner')
          .sort_values(['Country/Region','Date'])
)
merged
```

```python
merged.info()
```
