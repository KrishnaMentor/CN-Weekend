## What is Python?

1. It is a Programming Language used to communicate with machine.

2. It is beginner Friendly. It is Easy To understand , Best for All Data Related fields.

3. It's an open source. Free For All, And Base Code is Publicly 
   Available as a Documentation [Python.org](https://www.python.org/doc/).

4. It is Platform Independent (.byte code is generated) which can be used in any OS.

5. A Case Sensitive Language.

6. An Interpreted Language where each block of code is run and executed.

7. Rich in Libraries (Numpy, Pandas, MatplotLib, Seaborn, SQLAlchemy)

8. It is an Object Oriented Programming Language (Classes/Object).

9. Strong Community Support

```python
# Print <Object> Argument ['end' , 'sep']
# end -> '\n' [Next Line]
print("Coding")
print('Ninjas')
```

```python
# end -> " "
print("Coding" , end = " ")
print("Ninjas") # Coding Ninjas
```

```python
# end -> "\t" [Tab Space]
print("Coding" , end = "\t")
print("Ninjas") # Coding   Ninjas
```

```python
print("Coding", "Ninjas")
```

```python
print("Coding", "-" ,"Ninjas")
```

```python
# end = '\n'
print("Hello" , end = '\n')
print("World")
```
