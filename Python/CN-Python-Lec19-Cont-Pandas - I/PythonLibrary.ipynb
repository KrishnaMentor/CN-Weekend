```python
# Indexing , slicing , Operations on numpy array
arr_1d = np.array([11,22,33,44,55])
print(arr_1d[2]) # 33
print(arr_1d[-2]) # 44
print(arr_1d[-5]) # 11
# print(arr_1d[5]) # Index Error: Out of Range
print(arr_1d[0]) # 11
```

```python
# Shape of 3D Array (2,3,4) -> 2(depth),3(rows),4(cols)
arr_3d = np.array([
    [[11,22,33,44],
     [22,44,66,88],
     [11,33,55,77],],
    
    [[66,77,88,99],
     [11,55,77,99],
     [22,44,66,88],]
])
# indexing[depth,row,col]
print(arr_3d[-1,2,3]) # 88
print(arr_3d[-2,-3,-4]) # 11
# print(arr_3d[0,3,2]) # Error
# print(arr_3d[-2,0,4]) # Error
print(arr_3d[1,2,-3]) # 44
print(arr_3d[0,-1,-1]) # 77
print(arr_3d[-2,1,-4]) # 22
print(arr_3d[0,-3,3]) # 44
```

```python
# Slicing[Start = 0 , Stop = length [Exclusive], Step = 1 [default]]
arr_1d = np.array([11,22,33,44,55,66,77,88,99])
print(arr_1d[-3:]) # [77,88,99]
print(arr_1d[-3::-1]) # [77,66...11]
print(arr_1d[-3:9:-1]) # []
print(arr_1d[-3:-6:-1]) # [77,66,55]
print(arr_1d[-3:0:-1]) # [77,66....22]
print(arr_1d[5:9:2]) # [66,88]
print(arr_1d[5:4:2]) # []
print(arr_1d[5:4:-2]) # [66]
print(arr_1d[8:0:-2]) # [99,77,55,33]
print(arr_1d[-4::-2]) # [66,44,22]
print(arr_1d[3:8:-3]) # []
print(arr_1d[5:11:2]) # [66,88]
```

# Sorting An Array:
##### Sorting means arranging an elements in ascending (default) or descending order

```python
import numpy as np
arr_1d = np.array([77,11,22,55,33,44,99,66,88])
print(arr_1d)
print(np.sort(arr_1d))
```

```python
# Descending Order using slicing [::-1]
print(np.sort(arr_1d)[::-1]) 
```

```python
# Axis = 0 [Horizontal Axis][Rows] & Axis = 1 [Vertical Axis][Cols]
arr_2d = np.array([
    [11,22,33,44,55],
    [99,88,77,66,55],
    [11,33,55,77,99],
    [22,44,66,88,10],
    [11,15,17,19,21]
])
np.sort(arr_2d , axis = 0)
```

```python
# axis = 1 [Vertical Axis] [Col]
arr_2d = np.array([
    [11,22,33,44,55],
    [99,88,77,66,55],
    [11,33,55,77,99],
    [22,44,66,88,10],
    [11,15,17,19,21]
])
np.sort(arr_2d , axis = 1)
```

```python
# argsort [Indexing Sort]
arr_1d = np.array([55,11,44,33,22])
pos_arr = np.argsort(arr_1d)
print(arr_1d)
print(pos_arr)
```

```python
# Indexing 
arr_1d[1] # 11
```

```python
arr_1d[4] # 22
```

```python
arr_1d[3] # 33
```

```python
# Returns the indices that would sort an array. [argsort]
arr_1d[pos_arr]
```

```python
# argsort [Indexing Sort] [Descending]
arr_1d = np.array([55,11,44,33,22])
pos_arr = np.argsort(arr_1d)[::-1]
print(arr_1d)
print(pos_arr)
```

```python
arr_1d[pos_arr] # Descending
```

```python
# argsort -> 2D Matrix
# axis = 1 [Vertical] [Col]
arr_2d = np.array([
    [11,22,33,44,55],
    [99,88,77,66,55],
    [11,33,55,77,99],
    [22,44,66,88,10],
    [11,15,17,19,21]
])
print(arr_2d)
print(np.argsort(arr_2d , axis = 1))
```

```python
# How to Access an elements from a Series.
# .iloc[Positional Based Indexing][Non-Inclusive] [1,10)
# .loc [Labelled Based Indexing][Inclusive] [1,10]
data = [11,22,33,44,55,66,77,88,99]
series = pd.Series(data)
series
```

```python
print(series.iloc[-1]) # 99 # only .iloc can perform -ve indexing
print(series[6]) # +ve indexing are allowed # 77
print(series.loc[0]) # 11
print(series.iloc[-5]) # 55
# print(series.loc[-5]) # Error # KeyError: -5
print(series.loc[4]) # 55
```

```python
data = [11,22,33,44,55,66,77,88,99]
label = ['a','b','c','d','e','f','g','h','i']
series = pd.Series(data,label) # Positional Argument
print(series)
# print(series[0]) # KeyError
print(series['a']) # 11
print(series.iloc[-5]) # 55
print(series.iloc[7]) # 88
# print(series.loc[7]) # KeyError
print(series.loc['i']) # 99 
```

```python
# Slicing
data = [11,22,33,44,55,66,77,88,99]
label = ['a','b','c','d','e','f','g','h','i']
series = pd.Series(data,label) # Positional Argument
print(series)
print("\n Labelled Based Slicing using .loc")
print(series.loc['a':'e']) # [11,22,33,44,55]
print("\n Positional Based Slicing using .iloc")
print(series.iloc[0:5]) # [11,22,33,44,55]
```
