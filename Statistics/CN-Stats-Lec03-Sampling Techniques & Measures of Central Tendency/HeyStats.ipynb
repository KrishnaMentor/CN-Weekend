```python
# Hypothesis Testing [T-Test]
# H0 = Null Hypothesis - 95%
# H1 = Alternative Hypothesis - 5%
# Significance Level alpha (α), - 5% => p_value < 5% (α)
# df = degree of freedom (n-1) = sample size = 50 - 1 = 49
# p_value = probality value to either accept H0 or to reject H0
t_stat , p_value = stats.ttest_1samp(data,50)
print(t_stat)
print(p_value)
```

```python
# Sampling Technqiues
import numpy as np
import pandas as pd

df_population = pd.read_csv("Population_Survey_Data.csv")
df_population
```

```python
# Return a random sample of items from an axis of object.
# Use `random_state` for reproducibility. [np.random.seed(42)]

random_sample = df_population.sample(n=100, random_state = 42)
random_sample
```

```python
df_population['Region'].unique()
```

```python
df_population['Region'].value_counts()
```

```python
# Stratified Sampling -> Strata[Groups based on Demographics] -> random_sample
stratified_sample = (
    df_population.
    sample(frac = 1 , random_state = 11)
    .groupby('Region')
    .head(25)
)
stratified_sample
```

```python
stratified_sample['Region'].value_counts()
```

```python
stratified_sample[stratified_sample['Region'] == "East"]
```

```python
df_population['Education'].unique()
```

```python
df_population['Education'].value_counts()
```

```python
# Stratified Sampling -> Strata[Groups based on Demographics] -> random_sample
stratified_sample = (
    df_population.
    sample(frac = 0.5 , random_state = 11)
    .groupby('Education')
    .head(25)
)
stratified_sample
```

```python
stratified_sample['Education'].value_counts()
```

```python
# Clustered Sampling
df_manufacturing = pd.read_csv("Manufacturing_Data.csv")
df_manufacturing
```

```python
df_manufacturing['Defective'].unique()
```

```python
defective_prod = df_manufacturing[df_manufacturing['Defective'] == 1]
defective_prod
```

```python
defective_prod.shape
```

```python
defective_prod.shape[0] # 42 defective Products
```

```python
clusters= df_manufacturing['BatchNumber'].unique()
clusters
```

```python
selected_clusters = np.random.choice(clusters , size = 5 , replace=False)
selected_clusters
```

```python
cluster_sample = df_manufacturing[df_manufacturing['BatchNumber'].isin(selected_clusters)]
cluster_sample
```

```python
# Systematic Sampling => K = N/n -> Kth Factor -> Slicing [::k]
# Pop. Size = 1000 (N) , Sample Size -> 100(n) -> k = 1000/100 -> 10
systematic_sampling = df_population.iloc[::10 , :] # iloc[rows , cols]
systematic_sampling
```

```python
# Cluster Sampling for df_population
clusters = df_population['Age'].unique()
clusters
```

```python
selected_clusters = np.random.choice(clusters , size = 7, replace = False)
selected_clusters
```

```python
cluster_sample = df_population[df_population['Age'].isin(selected_clusters)]
cluster_sample
```

```python
# Statistical Analysis for Population 
mean_income = df_population['Income'].mean()
median_income = df_population['Income'].median()
mode_income = df_population['Income'].mode()
print(mean_income)
print(median_income)
print(mode_income[0])
```

```python
df_population['Income'].value_counts()
```

```python
df_population.describe()
```

```python
# Evaluating the Population Mean Vs Sample Mean from various Sampling Techniques

actual_mean_income = df_population['Income'].mean()
print(f"Actual Mean: {actual_mean_income}")

# Random Sample Technique
random_sample_mean_income = random_sample['Income'].mean()
print(f"Random Sample Mean: {random_sample_mean_income}")

# Stratified Sample Technique
stratified_sample_mean_income = stratified_sample['Income'].mean()
print(f"Stratified Sample Mean: {stratified_sample_mean_income}")

# Systematic Sample Technique
systematic_sample_mean_income = systematic_sampling['Income'].mean()
print(f"Systematic Sample Mean: {systematic_sample_mean_income}")

# Cluster Sample Technique
cluster_sample_mean_income = cluster_sample['Income'].mean()
print(f"Cluster Sample Mean: {cluster_sample_mean_income}")
```
