```python
# Scipy
import numpy as np
from scipy import stats
```

```python
dir(stats)
```

```python
# Generate Data
np.random.seed(11)
data = np.random.normal(loc=50,scale=10,size=1000)
data
```

```python
np.random.seed(42)
data = np.random.normal(loc=50,scale=10,size=1000)
data
```

```python
np.random.seed(42)
data = np.random.normal(loc=50,scale=10,size=1000)
data
```

```python
type(data)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
```

```python
print(mean)
print(median)
print(mode)
```

```python
# Hypothesis Testing [T-Test]
# H0 = Null Hypothesis - 95%
# H1 = Alternative Hypothesis - 5%
# Significance Level alpha (α), - 5% => p_value < 5% (α)
# df = degree of freedom (n-1) = sample size = 50 - 1 = 49
# p_value = probality value to either accept H0 or to reject H0
t_stat , p_value = stats.ttest_1samp(data,50)
print(t_stat)
print(p_value)
```

```python
# Scikit-Learn
# Data Normalization
from sklearn.preprocessing import StandardScaler
dir(StandardScaler)
```

```python
data = np.random.normal(loc=50,scale=10,size=1000)
data = data.reshape(-1,1)
data
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```

```python
mean = np.mean(scaled_data)
median = np.median(scaled_data)
std = np.std(scaled_data)

print(mean)
print(median)
print(std)
```

```python
# Polynomial Features
from sklearn.preprocessing import PolynomialFeatures
# X = np.random.rand(shape) 
X = np.random.rand(1000,1) * 10 # range (0,10)
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
X_poly
```

```python
X
```

```python
mean = np.mean(X_poly)
median = np.median(X_poly)
std = np.std(X_poly)
print(mean)
print(median)
print(std)
```

```python
# StatsModel
import pandas as pd
import statsmodels.api as sm
dir(sm.stats)
```

```python
data = np.random.normal(loc=50,scale=10,size=1000)
df = pd.DataFrame(data, columns = ['Value'])
df
```

```python
stats_model = sm.stats.DescrStatsW(df['Value'])
stats_model
```

```python
mean = stats_model.mean
median = df['Value'].median()
print(mean)
print(median)
```
